## 1 Basic function 


```python
def reverse_string(s):
    return s[::-1]

# Test it
print(reverse_string("hello"))  # 'olleh'
```

## IS Prime function

```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0: return False
    return True

# Test it
print(is_prime(17))  # True
```

```python
def is_prime(n):
    if n < 2:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True
```

### FIBONNACI Function

```python
def fibonacci(n):
    if n <= 1: return n
    a, b = 0, 1
    for _ in range(2, n+1):
        a, b = b, a + b
    return b

# Test it
print(fibonacci(10))  # 55
```

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)
```

## Sum of digit

```python
def digit_sum(n):
    return sum(int(d) for d in str(abs(n)))

# Test it
print(digit_sum(1234))  # 10
```

## Count distincs words

```python
def count_distinct_words(sentence):
    words = sentence.lower().split()
    return len(set(words))

# Test it
print(count_distinct_words("the cat sat on the mat"))  # 5
```
